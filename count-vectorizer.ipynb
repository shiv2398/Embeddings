```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Create a CountVectorizer object and fit it to your corpus of text documents
corpus = ["This is the first document",
         'this is the second sentence',
         'this is the second sentence ']
vectorizer = CountVectorizer()
vectorizer.fit_transform(corpus)
word_matrix=vectorizer.transform(corpus).toarray()
word_matrix
```

```python
word_df=pd.DataFrame(word_matrix,columns=vectorizer.get_feature_names())
word_df.head()
```

```python
# Create heatmap using seaborn
sns.heatmap(word_matrix, cmap='Blues', annot=True, xticklabels=vectorizer.get_feature_names())
plt.title("CountVectorizer Heatmap")
plt.show()
```

```python
total_counts = word_df.sum(axis=0)

# Create a bar chart using Matplotlib
plt.bar(vectorizer.get_feature_names(), total_counts)
plt.xticks(rotation=90)
plt.show()
```
